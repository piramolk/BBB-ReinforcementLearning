# Overview


## Prediction

 This type of task <b> predicts the expected total reward from any given state assuming the function (a|s) is given </b>
 
 ( in other words) Policy  is given, it calculates the Value function V with or without the model.
 
 
 ## Control 
 
This type of task finds the policy (a|s) that <b> maximizes the expected total reward from any given state.</b>
(in other words) Some Policy  is given , it finds the Optimal policy *.

# How to figure out V for unknown MDP ( assume we get the policy)?

## 1. Monte carlo approach

- Monte Carlo learning  it learns value functions directly from episodes of experience.
- Monte Carlo learning  We only get the reward at the end of an episode

    v(s) = E [Gt | St = s] and Gt = Rt+1+ Rt+2+

- we learn value functions from sample returns with the MDP
    - i.e. an average returns/rewards from the episode
    
 ### First visit MC

in this, we average returns only for first time s is visited in an episode


    
    
    
      
  
    
    
### Every visit MC
    
 
 
in this, we average returns for every time s is visited in an episode

# Temporal Difference

- Just like Monte Carlo  TD methods learn directly from episodes of experience and model free.
- Just like Dynamic Programming  TD methods boot strap , meaning it will not wait until the end of the episode to update the expected future reward estimation(V) , it will only wait until the next time step to update the value estimates.



These use Bellman equation to update the estimates


### Update Methods for V



- Gt  Monte Carlo Target
- Rt+1 +  V(St+1)  TD target
- Rt+1 +  V(St+1)  V(St)  TD error

example of update

        td_target = reward + gamma* V[next_state]
        td_error = td_target - V[current_state]
         V[current_state] += alpha*td_error

TD Target:
- R + gamma * Q(s',a')

TD Error
- R + gamma* Q(s',a') - Q(s,a)

if S' is terminal then Q(s',:)=0
